```python
# Import required libraries and dependencies
import pandas as pd
import hvplot.pandas
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from holoviews import opts
import holoviews as hv
import colorcet as cc
```

```python
# Load the data into a Pandas DataFrame
df_market_data = pd.read_csv(
    "Resources/crypto_market_data.csv",
    index_col="coin_id")

# Display sample data
df_market_data.head(10)
```

```python
# Generate summary statistics
df_market_data.describe()
```

```python
# Plot your data to see what's in your DataFrame
df_market_data.hvplot.line(
    width=800,
    height=400,
    rot=90
)
```

---

### Prepare the Data

```python
# Use the `StandardScaler()` module from scikit-learn to normalize the data from the CSV file
scaler = StandardScaler()

# Defining columns to normalize
columns_to_normalize = [
    'price_change_percentage_24h',
    'price_change_percentage_7d',
    'price_change_percentage_14d',
    'price_change_percentage_30d',
    'price_change_percentage_60d',
    'price_change_percentage_200d',
    'price_change_percentage_1y'
]
# Normalize the required columns
df_market_data[columns_to_normalize] = scaler.fit_transform(df_market_data[columns_to_normalize])
```

```python
# Create a DataFrame with the scaled data
scaled_df_market_data = df_market_data.copy()

# Copy the crypto names from the original data
crypto_names = df_market_data.index

# Set the coinid column as index
scaled_df_market_data['coin_id'] = crypto_names
scaled_df_market_data.set_index('coin_id', inplace=True)

# Display sample data
scaled_df_market_data.head()
```

---

### Find the Best Value for k Using the Original Data.

```python
# Create a list with the number of k-values from 1 to 11
k_values = list(range(1, 12))
```

```python
# Create an empty list to store the inertia values
inertia_values = []

# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
for k in k_values:
# 1. Create a KMeans model using the loop counter for the n_clusters
    kmeans = KMeans(n_clusters=k, random_state=42)
# 2. Fit the model to the data using `df_market_data_scaled`
    kmeans.fit(scaled_df_market_data)
# 3. Append the model.inertia_ to the inertia list
    inertia_values.append(kmeans.inertia_)
```

```python
# Create a dictionary with the data to plot the Elbow curve
elbow_data = {
    "k_values": k_values,
    "inertia_values": inertia_values
}

# Create a DataFrame with the data to plot the Elbow curve
elbow_df = pd.DataFrame(elbow_data)

elbow_df
```

```python
# Plot a line chart with all the inertia values computed with 
# the different values of k to visually identify the optimal value for k.
elbow_df.hvplot.line(x="k_values", y="inertia_values", title="Elbow Curve (Original Scaled Data)", xlabel="Number of Clusters (k)", ylabel="Inertia")
```

#### Answer the following question: 

**Question:** What is the best value for `k`?

**Answer:** 4

---

### Cluster Cryptocurrencies with K-means Using the Original Data

```python
# Initialise the K-Means model using the best value for k. According to the earlier Elbow Method, a cluster of 4 is the best fit for these data.
kmeans_model = KMeans(n_clusters=4, random_state=42)
```

```python
# Fit the K-Means model using the scaled data
kmeans_model.fit(scaled_df_market_data)
```

```python
# Predict the clusters to group the cryptocurrencies using the scaled data
clusters = kmeans_model.predict(scaled_df_market_data)

# Print the resulting array of cluster values.
print(clusters)
```

```python
# Create a copy of the DataFrame
clustered_scaled_df_market_data = scaled_df_market_data.copy()
```

```python
# Add a new column to the DataFrame with the predicted clusters
clustered_scaled_df_market_data['cluster'] = clusters

# Display sample data
clustered_scaled_df_market_data.head()
```

```python
# Create a scatter plot using hvPlot by setting 
# `x="price_change_percentage_24h"` and `y="price_change_percentage_7d"`. 
# Colour the graph points with the labels found using K-Means and 
# add the crypto name in the `hover_cols` parameter to identify 
# the cryptocurrency represented by each data point.

clustered_scaled_df_market_data.hvplot.scatter(
    x="price_change_percentage_24h",
    y="price_change_percentage_7d",
    by="cluster",  # Color points by cluster labels
    hover_cols=["coin_id"],  # Show coin_id on hover
    title="Crypto Clusters - Original Scaled Data"
)
```

### Optimise Clusters with Principal Component Analysis.

```python
# Create a PCA model instance and set `n_components=3`.
pca_model = PCA(n_components=3)
```

```python
# Use the PCA model with `fit_transform` to reduce to 
# three principal components.
principal_components = pca_model.fit_transform(scaled_df_market_data)
pca_df = pd.DataFrame(data=principal_components, columns=["PC1", "PC2", "PC3"], index=scaled_df_market_data.index)

# View the first five rows of the DataFrame. 
pca_df.head()
```

```python
# Retrieve the explained variance to determine how much information 
# can be attributed to each principal component.
explained_variance = pca_model.explained_variance_ratio_
print("Explained Variance for Each Principal Component:")
print(explained_variance)
```

#### Answer the following question: 

**Question:** What is the total explained variance of the three principal components?

**Answer:**  The total explained variance of the three principal components is approximately 0.895 or 89.5%. This means that these three principal components collectively capture 89.5% of the total variance in the original data.

### Find the Best Value for k Using the PCA Data

```python
# Create a list with the number of k-values from 1 to 11
pca_k_values = list(range(1, 12))
```

```python
# Create an empty list to store the inertia values
pca_inertia_values = []

# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
for k in pca_k_values:
# 1. Create a KMeans model using the loop counter for the n_clusters
    kmeans = KMeans(n_clusters=k, random_state=42)
# 2. Fit the model to the data using `df_market_data_scaled`
    kmeans.fit(pca_df)
# 3. Append the model.inertia_ to the inertia list
    pca_inertia_values.append(kmeans.inertia_)
```

```python
# Create a dictionary with the data to plot the Elbow curve
pca_elbow_data = {
    "k_values": pca_k_values,
    "inertia_values": pca_inertia_values
}

# Create a DataFrame with the data to plot the Elbow curve
pca_elbow_df = pd.DataFrame(pca_elbow_data)

pca_elbow_df
```

```python
# Plot a line chart with all the inertia values computed with 
# the different values of k to visually identify the optimal value for k.
pca_elbow_df.hvplot.line(x="k_values", y="inertia_values", title="Elbow Curve - PCA Data", xlabel="Number of Clusters (k)", ylabel="Inertia")
```

### Cluster Cryptocurrencies with K-means Using the PCA Data

```python
# Initialise the K-Means model using the best value for k. 4 was best deems as the best value for k.
kmeans_pca_model = KMeans(n_clusters=4, random_state=42)
```

```python
# Fit the K-Means model using the PCA data
kmeans_pca_model.fit(pca_df)
```

```python
# Predict the clusters to group the cryptocurrencies using the PCA data
pca_clusters = kmeans_pca_model.predict(pca_df)
# Print the resulting array of cluster values.
print(pca_clusters)
```

```python
# Create a copy of the DataFrame with the PCA data
clustered_pca_data_df = pca_df.copy()

# Add a new column to the DataFrame with the predicted clusters
clustered_pca_data_df['cluster'] = pca_clusters

# Display sample data
clustered_pca_data_df.head()
```

```python
# Create a scatter plot using hvPlot by setting 
# `x="PC1"` and `y="PC2"`. 
# Colour the graph points with the labels found using K-Means and 
# add the crypto name in the `hover_cols` parameter to identify 
# the cryptocurrency represented by each data point.

clustered_pca_data_df.hvplot.scatter(
    x="PC1",
    y="PC2",
    by="cluster",  # Color points by cluster labels
    hover_cols=["coin_id"],  # Show coin_id on hover
    title="Crypto Clusters - PCA Data"
)
```

### Visualise and Compare the Results

In this section, you will visually analyse the cluster analysis results by contrasting the outcome with and without using the optimisation techniques.

```python
# Define the original elbow curve plot
original_elbow_curve = elbow_df.hvplot.line(x="k_values", y="inertia_values", title="Elbow Curve Original Scaled Data vs PCA Data", xlabel="Number of Clusters (k)", ylabel="Inertia",
    label="Original Scaled Data"
).opts(color='blue')

# Define the PCA elbow curve plot
pca_elbow_curve = pca_elbow_df.hvplot.line(x="k_values", y="inertia_values", title="Elbow Curve Original Scaled Data vs PCA Data", xlabel="Number of Clusters (k)", ylabel="Inertia",
    label="PCA Data"
).opts(color='red')

# Overlay the two plots for comparison
composite_plot_elbow = (original_elbow_curve * pca_elbow_curve).opts(
    opts.Curve(width=600, height=300),
    opts.Layout(shared_axes=False)
)

# Display the composite plot of the elbow curve
composite_plot_elbow
```

```python

# Ensure that the color scheme for both clusters remain constant.
num_clusters = 4
cluster_colors = cc.glasbey_dark[:num_clusters] 

# Define the original elbow curve plot
original_cluster_image = clustered_scaled_df_market_data.hvplot.scatter(
    x="price_change_percentage_24h",
    y="price_change_percentage_7d",
    by="cluster",  
    hover_cols=["coin_id"],  
    title="Crypto Clusters Original Scaled Data (o) and PCA Data (x)",
    color=cluster_colors
)

# Define the PCA elbow curve plot
pca_cluster_image = clustered_pca_data_df.hvplot.scatter(
    x="PC1",
    y="PC2",
    by="cluster", 
    hover_cols=["coin_id"], 
    title="Crypto Clusters Original Scaled Data (o) and PCA Data (x)",
    marker="x",
    size=80,
    color=cluster_colors
    
)

# Overlay the two plots for comparison
combined_cluster = (original_cluster_image * pca_cluster_image).opts(
    opts.Scatter(width=400, height=300),
    opts.Layout(shared_axes=False),
)

# Display the composite plot of the combined clusters
combined_cluster
```
